4. –ü–æ—Å—á–∏—Ç–∞—Ç—å —á–∞—Å—Ç–æ—Ç–Ω–æ—Å—Ç—å –∫–∞–∂–¥–æ–≥–æ —Å–ª–æ–≤–∞ –≤ –∫–∞–∂–¥–æ–º –∏–∑ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤;

#### –°–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–∞—è —Ä–µ–∞–ª–∏–∑–∞—Ü–∏—è

–ü–æ–ø—Ä–æ–±—É–µ–º —Ä–µ–∞–ª–∏–∑–æ–≤–∞—Ç—å –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏—é –º–µ—Ç–æ–¥–æ–º "–º–µ—à–æ–∫ —Å–ª–æ–≤" —Å–∞–º–æ—Å—Ç–æ—è—Ç–µ–ª—å–Ω–æ.

–ò–º–ø–æ—Ä—Ç–∏—Ä—É–µ–º –Ω–µ–æ–±—Ö–æ–¥–∏–º—ã–µ –º–æ–¥—É–ª–∏:

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from collections import defaultdict
```

–î–æ–±–∞–≤–ª—è–µ–º –ø—Ä–∏–º–µ—Ä—ã –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ (–ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–π):

```python
data = ['She loves pizza, pizza is delicious.','She is good person.','Good people are the best.']
```

–î–µ–ª–∞–µ–º –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫—É –∏ –∑–∞–ø–∏—Å—ã–≤–∞–µ–º —Å–ª–æ–≤–∞ –≤ —Å–ª–æ–≤–∞—Ä—å:

```python
sentences = [] # —Ç–æ–∫–µ–Ω–∏–∑–∏—Ä–æ–≤–∞–Ω–Ω—ã–µ –¥–æ–∫—É–º–µ–Ω—Ç—ã
vocab = [] # —Å–ª–æ–≤–∞—Ä—å
for sentence in data:
    x = word_tokenize(sentence) # —Ç–æ–∫–µ–Ω–∏–∑–∏—Ä—É–µ–º –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è
    sentence = [w.lower() for w in x if w.isalpha()] # —É–±–∏—Ä–∞–µ–º –∑–Ω–∞–∫–∏ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è –∏ –ø—Ä–∏–≤–æ–¥–∏–º –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
    sentences.append(sentence)
    for word in sentence:
        if word not in vocab:
            vocab.append(word)

len_vector = len(vocab) # –æ–ø—Ä–µ–¥–µ–ª–∏–º —á–∏—Å–ª–æ —Å–ª–æ–≤ –≤ —Å–ª–æ–≤–∞—Ä–µ
```

üìå –ù–µ–æ–±—Ö–æ–¥–∏–º–æ –ø—Ä–∏–ø–∏—Å–∞—Ç—å –∫–∞–∂–¥–æ–º—É —Å–ª–æ–≤—É –∏–Ω–¥–µ–∫—Å. –ù–∞ –≤—ã—Ö–æ–¥–µ –Ω—É–∂–Ω–æ –ø–æ–ª—É—á–∏—Ç—å —Å–ª–æ–≤–∞—Ä—å, –≥–¥–µ –∫–ª—é—á–∞–º–∏ —è–≤–ª—è—é—Ç—Å—è —Å–ª–æ–≤–∞, –∞ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ ‚Äî –∏—Ö –∏–Ω–¥–µ–∫—Å—ã –Ω–∞—á–∏–Ω–∞—è —Å 0.

```python
index_word = {} # —Å–æ–∑–¥–∞–µ–º –ø—É—Å—Ç–æ–π —Å–ª–æ–≤–∞—Ä—å
i = 0
for word in vocab:
  # –ú–µ—Å—Ç–æ –¥–ª—è –≤–∞—à–µ–≥–æ –∫–æ–¥–∞
index_word
```

–¢–µ–ø–µ—Ä—å –Ω–∞–ø–∏—à–µ–º —Ñ—É–Ω–∫—Ü–∏—é –¥–ª—è –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏–∏:

```python
def bag_of_words(sentence):
    count_dict = defaultdict(int) # —Å–æ–∑–¥–∞–µ–º —Å–ª–æ–≤–∞—Ä—å —Å–æ –∑–Ω–∞—á–µ–Ω–∏—è–º–∏ –ø–æ —É–º–æ–ª—á–∞–Ω–∏—é
    vector = np.zeros(len_vector) # —Å–æ–∑–¥–∞–µ–º –≤–µ–∫—Ç–æ—Ä –ø–æ –¥–ª–∏–Ω–µ —Å–ª–æ–≤–∞—Ä—è, –∑–∞–ø–æ–ª–Ω–µ–Ω–Ω—ã–π –Ω—É–ª—è–º–∏
    for item in sentence:
        count_dict[item] += 1 # —Å—á–∏—Ç–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—Ö–æ–∂–¥–µ–Ω–∏–π —Å–ª–æ–≤–∞ –≤ –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏–∏
    for key, item in count_dict.items():
        vector[index_word[key]] = item # –∑–∞–ø–∏—Å—ã–≤–∞–µ–º –∫–æ–ª–∏—á–µ—Å—Ç–≤–æ –≤—Ö–æ–∂–¥–µ–Ω–∏–π —Å–ª–æ–≤–∞ –≤ –≤–µ–∫—Ç–æ—Ä –ø–æ –µ–≥–æ –∏–Ω–¥–µ–∫—Å—É
    return vector
```

–ü—Ä–µ–æ–±—Ä–∞–∑—É–µ–º –ø—Ä–µ–¥–ª–æ–∂–µ–Ω–∏—è –≤ –≤–µ–∫—Ç–æ—Ä—ã:

```python
vectors = []
for sentence in sentences:
  vectors.append(bag_of_words(sentence))
```

–ü—Ä–µ–¥—Å—Ç–∞–≤–∏–º —Ä–µ–∑—É–ª—å—Ç–∞—Ç –≤ –≤–∏–¥–µ –¥–∞—Ç–∞—Ñ—Ä–µ–π–º–∞:

```python
bow = {data[0]: vectors[0], data[1]: vectors[1], data[2]: vectors[2]}
df1 = pd.DataFrame.from_dict(bow, orient='index', columns = index_word.keys())
df1
```

### TF-IDF

TF-IDF (TF ‚Äî term frequency, IDF ‚Äî inverse document frequency)  ‚Äî —ç—Ç–æ —Å–ø–æ—Å–æ–± –≤–µ–∫—Ç–æ—Ä–∏–∑–∞—Ü–∏–∏ —Ç–µ–∫—Å—Ç–∞, –æ—Ç—Ä–∞–∂–∞—é—â–∏–π –≤–∞–∂–Ω–æ—Å—Ç—å —Å–ª–æ–≤–∞ –≤ –¥–æ–∫—É–º–µ–Ω—Ç–µ, –∞ –Ω–µ —Ç–æ–ª—å–∫–æ —á–∞—Å—Ç–æ—Ç—É –µ–≥–æ –ø–æ—è–≤–ª–µ–Ω–∏—è.

–ß–∞—Å—Ç–æ—Ç–∞ —Å–ª–æ–≤ (TF) ‚Äî —ç—Ç–æ –º–µ—Ä–∞ —á–∞—Å—Ç–æ—Ç—ã —É–ø–æ—Ç—Ä–µ–±–ª–µ–Ω–∏—è —Å–ª–æ–≤–∞ $w$ –≤ –¥–æ–∫—É–º–µ–Ω—Ç–µ $d$. TF –æ–ø—Ä–µ–¥–µ–ª—è–µ—Ç—Å—è –∫–∞–∫ –æ—Ç–Ω–æ—à–µ–Ω–∏–µ –ø–æ—è–≤–ª–µ–Ω–∏—è —Å–ª–æ–≤–∞ –≤ –¥–æ–∫—É–º–µ–Ω—Ç–µ –∫ –æ–±—â–µ–º—É –∫–æ–ª–∏—á–µ—Å—Ç–≤—É —Å–ª–æ–≤ –≤ –¥–æ–∫—É–º–µ–Ω—Ç–µ.

$$TF(w,d) = \frac{–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ\:–≤—Ö–æ–∂–¥–µ–Ω–∏–π\:—Å–ª–æ–≤–∞\:w\:–≤\:–¥–æ–∫—É–º–µ–Ω—Ç–µ\:d}{–æ–±—â–µ–µ\:–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ\:—Å–ª–æ–≤\:n\:–≤\:–¥–æ–∫—É–º–µ–Ω—Ç–µ\:d}$$

–û–±—Ä–∞—Ç–Ω–∞—è —á–∞—Å—Ç–æ—Ç–∞ –¥–æ–∫—É–º–µ–Ω—Ç–æ–≤ (IDF) ‚Äî  —ç—Ç–æ –º–µ—Ä–∞ –≤–∞–∂–Ω–æ—Å—Ç–∏ —Å–ª–æ–≤–∞. –ù–µ–∫–æ—Ç–æ—Ä—ã–µ —Å–ª–æ–≤–∞ –º–æ–≥—É—Ç –ø—Ä–∏—Å—É—Ç—Å—Ç–≤–æ–≤–∞—Ç—å –Ω–∞–∏–±–æ–ª–µ–µ —á–∞—Å—Ç–æ, –Ω–æ –Ω–µ –∏–º–µ—é—Ç –±–æ–ª—å—à–æ–≥–æ –∑–Ω–∞—á–µ–Ω–∏—è. IDF –ø—Ä–∏—Å–≤–∞–∏–≤–∞–µ—Ç –≤–µ—Å –∫–∞–∂–¥–æ–º—É —Å–ª–æ–≤—É –≤ –∑–∞–≤–∏—Å–∏–º–æ—Å—Ç–∏ –æ—Ç –µ–≥–æ —á–∞—Å—Ç–æ—Ç—ã –≤ –∫–æ—Ä–ø—É—Å–µ $D$.

$$IDF(w,D) = ln(\frac{–æ–±—â–µ–µ\:–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ\:–¥–æ–∫—É–º–µ–Ω—Ç–æ–≤\:N\:–≤\:–∫–æ—Ä–ø—É—Å–µ\:D}{–∫–æ–ª–∏—á–µ—Å—Ç–≤–æ\:–¥–æ–∫—É–º–µ–Ω—Ç–æ–≤,\:—Å–æ–¥–µ—Ä–∂–∞—â–∏—Ö\:—Å–ª–æ–≤–æ\:w})$$

TF-IDF —è–≤–ª—è–µ—Ç—Å—è –ø—Ä–æ–∏–∑–≤–µ–¥–µ–Ω–∏–µ–º TF –∏ IDF.
$$TF-IDF(w,d,D)=TF(w,d)*IDF(w,D)$$
